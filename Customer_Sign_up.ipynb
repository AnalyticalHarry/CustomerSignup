```python
import pandas as pd 
import numpy as np 
import openpyxl
import os
import warnings
warnings.filterwarnings("ignore")
```

```python
def signup_process(case):
    #Greeting to user
    print("Hello, I heard you are interested in our product.") 

    #List are used to store value
    Fname_list=[]; #First Name
    Lname_list=[]; #Last Name
    Country_list=[]; #Country
    Address_list=[]; #First line of Address
    Postcode_list=[]; #Postcode
    Gender_list=[]; #Sex
    Email_list=[]; #Email Address
    Contact_list=[] #Contact Number

    for i in range(case): #Iterate in range of case, which was total number of people appear for registration
        #for loop interating and assigning value to list
        value=input("Enter your First Name:  ").title()
        value1=input("Enter your Last Name:  ").title()
        value2=input("Enter your Country :  ").title()
        value3=input("Enter your Address:  ").title()
        value4=input("Enter your Postcode :  ").title()
        value5=input("Gender:  ").title()
        value6=input("Enter your Email:  ").lower()
        value7=input("Enter your Contact Number:  ")
        print()
        print(f"Sucessfully, sign-up completed:", i+1)
        print()
        Fname_list.append(value)
        Lname_list.append(value1)
        Country_list.append(value2)
        Address_list.append(value3)
        Postcode_list.append(value4)
        Gender_list.append(value5)
        Email_list.append(value6)
        Contact_list.append(value7)
        #assigning value to key inside the dictonary
    dict = {"first_name":Fname_list,
                "last_name":Lname_list,
                "country":Country_list,
                "address":Address_list,
                "postcode":Postcode_list, 
                "gender":Gender_list,
                "email":Email_list ,
                "contact":Contact_list,}
    df_sample = pd.DataFrame(dict)
    if os.path.exists(r'C:\Users\Public\Downloads\data.xlsx'):
        df = pd.read_excel(r'C:\Users\Public\Downloads\data.xlsx')
        df = df.append(df_sample, ignore_index=True)
    else:
        df = df_sample

    writer = pd.ExcelWriter(r'C:\Users\Public\Downloads\data.xlsx')
    df.to_excel(writer, index=False)
    writer.save()
    writer.close()



    
def update_record(email, new_first_name, new_last_name, new_country, new_address, new_postcode, new_gender, new_contact):
    if os.path.exists(r'C:\Users\Public\Downloads\data.xlsx'):
        df = pd.read_excel(r'C:\Users\Public\Downloads\data.xlsx')
        df.loc[df['email'] == email, 'first_name'] = new_first_name
        df.loc[df['email'] == email, 'last_name'] = new_last_name
        df.loc[df['email'] == email, 'country'] = new_country
        df.loc[df['email'] == email, 'address'] = new_address
        df.loc[df['email'] == email, 'postcode'] = new_postcode
        df.loc[df['email'] == email, 'gender'] = new_gender
        df.loc[df['email'] == email, 'contact'] = new_contact
        writer = pd.ExcelWriter(r'C:\Users\Public\Downloads\data.xlsx')
        df.to_excel(writer, index=False)
        writer.save()
        writer.close()
    else:
        print("The data file does not exist.")


def delete_record(email):
    if os.path.exists(r'C:\Users\Public\Downloads\data.xlsx'):
        df = pd.read_excel(r'C:\Users\Public\Downloads\data.xlsx')
        matching_records = df[df['email'] == email]
        if len(matching_records) == 0:
            print("Error: Email not found.")
            return
        elif len(matching_records) > 1:
            print("Multiple records found with the same email:")
            for i, record in matching_records.iterrows():
                print(f"{i}: {record['first_name']} {record['last_name']}, {record['country']}, {record['address']}, {record['postcode']}, {record['gender']}, {record['contact']}")
            choice = int(input("Enter the index number of the record you want to delete: "))
            df = df[df.index != choice]
        else:
            df = df[df.email != email]
        writer = pd.ExcelWriter(r'C:\Users\Public\Downloads\data.xlsx')
        df.to_excel(writer, index=False)
        writer.save()
        writer.close()
        print("Record deleted successfully.")
    else:
        print("The data file does not exist.")


    
def main():
    try:
        case = int(input("How many people are here to sign-up? (for example 1,2,3....,n or 0 for exit):  ")) #total number of people for registration
        if case != 0:
            signup_process(case)
    except ValueError:
        print("Error: Please enter a valid number of people.")

    while True:
        print("\nMenu:")
        print("1. Type 'up' to Update a record")
        print("2. Type 'del' to Delete a record")
        print("3. Type 'exit' to Exit")
        choice = input("Enter your choice: ")

        if choice == 'update':
            email = input("Enter the email of the record to update: ")
            if email == '':
                print("Error: Email cannot be empty. Please enter a valid email.")
                continue
            new_first_name = input("Enter the new first name: ")
            if new_first_name == '':
                print("Error: First name cannot be empty. Please enter a valid first name.")
                continue
            new_last_name = input("Enter the new last name: ")
            if new_last_name == '':
                print("Error: Last name cannot be empty. Please enter a valid last name.")
                continue
            new_country = input("Enter the new country: ")
            if new_country == '':
                print("Error: Country cannot be empty. Please enter a valid country.")
                continue
            new_address = input("Enter the new address: ")
            if new_address == '':
                print("Error: Address cannot be empty. Please enter a valid address.")
                continue
            new_postcode = input("Enter the new postcode: ")
            if new_postcode == '':
                print("Error: Postcode cannot be empty. Please enter a valid postcode.")
                continue
            new_gender = input("Enter the new gender: ")
            if new_gender == '':
                print("Error: Gender cannot be empty. Please enter a valid gender.")
                continue
            new_contact = input("Enter the new contact: ")
            try:
                new_contact = int(new_contact)
            except ValueError:
                print("Error: Contact must be an integer. Please enter a valid contact.")
                continue
            update_record(email, new_first_name, new_last_name, new_country, new_address, new_postcode, new_gender, new_contact)
        elif choice == 'del':
            email = input("Enter the email of the record to delete: ")
            if email == '':
                print("Error: Email cannot be empty. Please enter a valid email.")
                continue
            delete_record(email)
        elif choice == 'exit':
            break
        else:
            print("Error: Please enter a valid choice.")

if __name__ == "__main__":
    main()
```
